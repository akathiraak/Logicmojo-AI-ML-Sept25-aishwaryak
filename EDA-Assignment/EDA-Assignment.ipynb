```python
# Imputing Name_Type_Suite with mode values to fix the null values

df1.groupby(['NAME_TYPE_SUITE'])['NAME_TYPE_SUITE'].count().plot.bar()
```

```python

#Since most of the values are 'Unaccompanied' we would use it to fill all the null values and procees with our analysis thereafter.

#Also it tells us that most of the loan applications were from those people who were Unaccompanied
# Imputing Values

mod = df1.NAME_TYPE_SUITE.mode()
mod
```

```python
df1.loc[df1.NAME_TYPE_SUITE.isnull(), 'NAME_TYPE_SUITE'] = 'Unaccompanied'

df1.isnull().sum()
```

```python
# Checking if we can replace 2 null values in CNT_FAM_MEMBERS by any obvious value from the graph

df1.groupby(['CNT_FAM_MEMBERS'])['CNT_FAM_MEMBERS'].count().plot.bar()
```

```python
#Imputing the null values in CNT_FAM_MEMBERS

df1.loc[df1.CNT_FAM_MEMBERS.isnull(), 'CNT_FAM_MEMBERS'] = 2.0

df1.isnull().sum()
```

```python
#Now that we see all the required columns are null - free, we need to check ahead if they have any more trash values to be dealth with, 
#if not we would proceed with our analysis

#Classifying the columns in category or numeric


numeric_col=['TARGET','CNT_CHILDREN','AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE','REGION_POPULATION_RELATIVE','DAYS_BIRTH',
                'DAYS_EMPLOYED','DAYS_REGISTRATION','DAYS_ID_PUBLISH','HOUR_APPR_PROCESS_START','LIVE_REGION_NOT_WORK_REGION', 'REG_CITY_NOT_LIVE_CITY',
       'REG_CITY_NOT_WORK_CITY', 'LIVE_CITY_NOT_WORK_CITY']

categoric_col =['SK_ID_CURR', 'NAME_CONTRACT_TYPE', 'CODE_GENDER','FLAG_OWN_CAR', 'FLAG_OWN_REALTY',
                'NAME_TYPE_SUITE','NAME_INCOME_TYPE','NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','NAME_HOUSING_TYPE','CNT_FAM_MEMBERS',
               'REGION_RATING_CLIENT','WEEKDAY_APPR_PROCESS_START','ORGANIZATION_TYPE']
```
